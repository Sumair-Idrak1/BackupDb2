# **Lab Task 1**

## Write a Python code for the following scenario:

## Suppose an instructor of a class wants to maintain record of his students in a python dictionary data structure. The designed system asks the instructor for a record of each student one by one as a console input and passes it through a set of sequential steps before generating a dictionary data structure.

## Write a Python script to perform following set of operations to generate dictionary of data for students.

## 1. The script should ask the instructor repeatedly for a “student ID”, “name” and his “residential city name” and store it in three different lists. If the instructor types “stop” the loop should exit.

## 2. Traverse the three lists generated in “part 1” and create a dictionary data structure out of it. The script should take ID of student as a Key of the dictionary and rest of the attributes as a list within that dictionary in following sequence: {1:[“name”,”city”]}

## Once dictionary is created, the instructor should be able to query the data. He should provide ID of the student as a console input and script should return the name and city of the student.

```python
# Initialize empty lists to store student data
student_ids = []
student_names = []
student_cities = []

# Part 1: Collect student data
while True:
    student_id = input("Enter student ID (or type 'stop' to finish): ")

    if student_id.lower() == 'stop':
        break

    student_name = input("Enter student name: ")
    student_city = input("Enter student's residential city: ")

    # Append the input values to the respective lists
    student_ids.append(student_id)
    student_names.append(student_name)
    student_cities.append(student_city)
```

### Create a dictionary with student data

```python
students_dict = {}

for i in range(len(student_ids)):
    students_dict[student_ids[i]] = [student_names[i], student_cities[i]]

# Print out the generated dictionary (optional)
print("\nStudent Data Dictionary:", students_dict)
```

### Allow the instructor to query the dictionary

```python
while True:
    query_id = input("\nEnter student ID to query (or type 'stop' to exit): ")

    if query_id.lower() == 'stop':
        print("Exiting the program.")
        break

    # Check if the student ID exists in the dictionary
    if query_id in students_dict:
        name, city = students_dict[query_id]
        print(f"Student ID: {query_id} - Name: {name}, City: {city}")
    else:
        print("Student ID not found.")
```

## BREADTH FIRST SEARCH

```python
graph = {'A':['B','C','D'],
         'B':['E','F'],
         'C':['G','H'],
         'D':['I','J'],
         'E':['K','L'],
         'F':['L','M'],
         'G':['N'],
         'H':['O','P'],
         'I':['P','Q'],
         'J':['R'],
         'K':['S'],
         'L':['T'],
         'P':['U'],
         'S':[],
         'T':[],
         'M':[],
         'N':[],
         'O':[],
         'U':[],
         'Q':[],
         'R':[]}

def bfs_traversal(graph, start, goal):

    opened = [start]
    closed = []

    while opened:
        #Remove leftmost child from the opened list and call it node
        node = opened.pop(0)
        #If node is goal then return SUCCESS alogn with the traversal sequence.
        if node == goal:
          closed.append(node)
          return "SUCCESS", closed
        else:
            #Add node to closed
            closed.append(node)
            #Generate children of node. If they are already present in opened, discard them, else push them on right end of opened
            opened = opened + [child for child in graph[node] if child not in opened and child not in closed]
    return 'GOAL Not FOUND'

msg, bfsTraversal = bfs_traversal(graph,'A','M')
print(msg,"\nBFS Traversal = ", bfsTraversal)
```

## DEPTH FIRST SEARCH

```python
graph = {'A':['B','C','D'],
         'B':['E','F'],
         'C':['G','H'],
         'D':['I','J'],
         'E':['K','L'],
         'F':['L','M'],
         'G':['N'],
         'H':['O','P'],
         'I':['P','Q'],
         'J':['R'],
         'K':['S'],
         'L':['T'],
         'P':['U'],
         'S':[],
         'T':[],
         'M':[],
         'N':[],
         'O':[],
         'U':[],
         'Q':[],
         'R':[]}

def dfs_traversal(graph, start, goal):

    opened = [start]
    closed = []
    while opened:
        node = opened.pop(0)
        if node == goal:
          closed.append(node)
          return "Goal Reached", closed
        else:
          closed.append(node)
          opened = [child for child in graph[node] if child not in opened and child not in closed] + opened
    return 'GOAL Not FOUND'

msg, dfsTraversal = dfs_traversal(graph,'A','M')
print(msg,"\nDFS Traversal = ", dfsTraversal)
```

### Step 2: BFS and DFS Functions

```python
from collections import deque

def bfs(graph, start, goal):
    visited = set()
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        city = path[-1]

        if city == goal:
            return path

        if city not in visited:
            visited.add(city)
            for neighbor in graph.get(city, []):
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    return None
```

```python
def dfs(graph, start, goal, path=None, visited=None):
    if visited is None:
        visited = set()
    if path is None:
        path = [start]

    visited.add(start)

    if start == goal:
        return path

    for neighbor in graph.get(start, []):
        if neighbor not in visited:
            new_path = dfs(graph, neighbor, goal, path + [neighbor], visited)
            if new_path:
                return new_path
    return None
```

## 1. Add functionality to detect diagonal conflicts in the board to evaluate board fitness.
## 2. Add mutation operation in the above code as a separate function similar to the ones added for crossover, fitness and other operations of Genetic Algorithms above. Generate a random number in the range 0 to 7 to decide about which Queen (Gene) to replace in the chromosome with the randomly generated number in the range (0 to 7) representing column number.
## 3. Restructure the complete code to take the number of boards (population size) from the user as an input and perform all the operations of Genetic Algorithm.

```python
import numpy as np
import random
# Function to create a random 8-queens board
def create_8_queens_board():
    queens_board = np.full((8,8), '.')  # Create an 8x8 board filled with '.'
    for row in range(8):
        col = random.randint(0, 7)  # Randomly choose a column for the queen in this row
        queens_board[row, col] = 'Q'  # Place the queen ('Q') on the board
    return queens_board  # Return the board


# Function to extract chromosome from board
def board_chromosome(board):
    queens_positions = np.argwhere(board == 'Q')
    cols = queens_positions[:, 1]
    return cols

# Function to calculate fitness including diagonal conflicts
def calculate_fitness(chromosome):
    fitness = 28  # Maximum non-attacking pairs (8*7/2)

    # Check column conflicts
    unique_cols, counts = np.unique(chromosome, return_counts=True)
    col_conflicts = sum([c*(c-1)//2 for c in counts])
    fitness -= col_conflicts

    # Check diagonal conflicts
    diagonal_conflicts = 0
    for i in range(len(chromosome)):
        for j in range(i+1, len(chromosome)):
            if abs(i - j) == abs(chromosome[i] - chromosome[j]):
                diagonal_conflicts += 1
    fitness -= diagonal_conflicts

    return fitness

# Function to select parents using tournament selection
def selection(population, fitness_values, num_parents=2):
    combined = list(zip(fitness_values, population))
    parents = []
    for _ in range(num_parents):
        candidates = random.sample(combined, min(3, len(combined)))
        best = max(candidates, key=lambda x: x[0])
        parents.append(best[1])
    return parents

# Function to perform single-point crossover
def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1)-1)
    offspring1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    offspring2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
    return offspring1, offspring2

# Function to perform mutation (Lab Task 2)
def mutation(chromosome, mutation_rate=0.1):
    if random.random() < mutation_rate:
        gene_to_mutate = random.randint(0, len(chromosome)-1)
        new_value = random.randint(0, 7)
        chromosome[gene_to_mutate] = new_value
    return chromosome

# Function to create new generation
def create_new_generation(population, fitness_values, mutation_rate=0.1):
    new_population = []
    best_idx = np.argmax(fitness_values)
    new_population.append(population[best_idx])

    while len(new_population) < len(population):
        parents = selection(population, fitness_values, 2)
        offspring1, offspring2 = crossover(parents[0], parents[1])
        offspring1 = mutation(offspring1, mutation_rate)
        offspring2 = mutation(offspring2, mutation_rate)
        new_population.extend([offspring1, offspring2])

    return new_population[:len(population)]

# Main Genetic Algorithm function (Lab Task 4 - takes population size as input)
def genetic_algorithm_8queens():
    population_size = int(input("Enter population size (number of boards): "))
    max_generations = 100
    mutation_rate = 0.1

    # Initialize population
    population = []
    for _ in range(population_size):
        board = create_8_queens_board()
        chromosome = board_chromosome(board)
        population.append(chromosome)

    best_solution = None
    best_fitness = -1

    for generation in range(max_generations):
        fitness_values = [calculate_fitness(chrom) for chrom in population]
        current_best = max(fitness_values)

        if current_best == 28:  # Found solution
            best_idx = np.argmax(fitness_values)
            best_solution = population[best_idx]
            print(f"\nSolution found in generation {generation}:")
            print_board(best_solution)
            return

        if current_best > best_fitness:
            best_fitness = current_best
            best_idx = np.argmax(fitness_values)
            best_solution = population[best_idx]

        population = create_new_generation(population, fitness_values, mutation_rate)

        if generation % 10 == 0:
            print(f"Generation {generation}: Best fitness = {best_fitness}")

    print(f"\nBest solution found (fitness = {best_fitness}):")
    print_board(best_solution)

# Helper function to print the board
def print_board(chromosome):
    board = np.full((8,8), '.')
    for row, col in enumerate(chromosome):
        board[row, col] = 'Q'
    print(board)

# Run the genetic algorithm
print("8-Queens Problem using Genetic Algorithm")
print("----------------------------------------")
genetic_algorithm_8queens()
```

##Apply Feature Creation preprocessing step on the Titanic dataset to create a Family Size feature which calculates Family Size for each Passenger using following Equation. FamilySize = SibSp + Parch + 1

##Apply ColumnTransformer, FunctionTransformer and Sklearn Pipeline on the Titanic dataset.

##Use ColumnTransformer, FunctionTransformer and Sklearn Pipeline features to preprocess the following dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data = pd.read_csv('/content/drive/MyDrive/titanic.csv')
```

```python
data.head(5)
```

```python
# Define function to create FamilySize feature
def create_family_size(X):
    # Convert input to DataFrame if it's a numpy array
    if isinstance(X, np.ndarray):
        X = pd.DataFrame(X, columns=['SibSp', 'Parch'])
    X['FamilySize'] = X['SibSp'] + X['Parch'] + 1
    return X[['FamilySize']]  # Return only the new column
```

```python
# Create FunctionTransformer
family_size_transformer = FunctionTransformer(create_family_size)
```

```python
# Get all original columns
original_columns = data.columns.tolist()
```

```python
# Create ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('family_size', family_size_transformer, ['SibSp', 'Parch'])
    ],
    remainder='passthrough'
)
```

```python
# Create pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor)
])
```

```python
# Apply the pipeline
data_processed = pipeline.fit_transform(data)
```

```python
# Get the new column names
# ColumnTransformer puts the transformed columns first, then the remainder
processed_columns = ['FamilySize'] + [col for col in original_columns if col not in ['SibSp', 'Parch']]
```

```python
data.head(5)
```

```python
# Convert back to DataFrame
df_processed = pd.DataFrame(data_processed, columns=processed_columns)

print(df_processed.head())
```
